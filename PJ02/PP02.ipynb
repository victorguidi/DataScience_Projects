```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
AAPL = yf.Ticker("AAPL")

# show actions (dividends, splits)
AAPL.actions

# show dividends
AAPL.dividends

# show splits
AAPL.splits
```

```python
stock = yf.download('AAPL') #Getting the data from Apple
```

```python
#Creating the DataFrame
df = pd.DataFrame(stock, columns=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Target'])

df
```

```python
df['Target'].iloc[0] = 0 #Starting the target with 0 for the first value

df
```

```python
#Creating the loop  that will create the target value for each day
#1 = Adj close was higher than the day before
#0 = Adj close was lower than the day before

i= 1
j = 0

while True:
    
    if(df.iat[i, 4] -  df.iat[j, 4] > 0):
        df.iat[i, 6] = 1
        i = i + 1
        j = j + 1
    else:
        df.iat[i, 6] = 0
        i = i + 1
        j = j + 1
    
    if i == 10289:
        break
```

```python
df
```

```python
#Using SkLearn to build the algorithm, in this case we are going to use Random Forest based on Gaussian Classifier

X = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] #Features
y = df[['Target']] #Labels

#Time to split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3) #70% training and 30% testing
```

```python
#Create a Guassian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train, y_train.values.ravel())

y_pred=clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
